```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
filename = "../data/tra_data.pickle"
with open(filename, 'rb') as handle:
    data = pickle.load(handle)
    
x_tra = data["x_tra"]
u_tra = data["u_tra"]
```

```python
time = np.arange(625) * 1/125
x_tra.shape, u_tra.shape, time.shape
```

```python
fig, axs = plt.subplots(5, 1, figsize=(10, 15), sharex=True)
lw = 0.2

for i in range(x_tra.shape[1]):
    axs[0].plot(time, x_tra[:, i, 0, 0], color="cornflowerblue", linewidth=lw)
    axs[1].plot(time, x_tra[:, i, 1, 0], color="cornflowerblue", linewidth=lw)
    axs[2].plot(time, x_tra[:, i, 2, 0], color="cornflowerblue", linewidth=lw)
    axs[3].plot(time, x_tra[:, i, 3, 0], color="cornflowerblue", linewidth=lw)
    axs[4].plot(time, u_tra[:, i, 0, 0], color="cornflowerblue", linewidth=lw)

ylabels = ["Cart Position [m]", "Pole Angle [rad]", "Cart Velocity [m/s]", "Pole Angular Velocity [rad/s]", "Control [N]"]
for i in range(5):
    axs[i].axhline(0.0, color="darkorange", linestyle="dashed", zorder=-20)
    axs[i].set_ylabel(ylabels[i], labelpad=0.1, fontsize=12)
    axs[i].yaxis.set_label_coords(-0.08, 0.5)

axs[1].axhline(np.pi, color="black", linestyle="dashed", zorder=-20)
axs[1].axhline(-np.pi, color="black", linestyle="dashed", zorder=-20)
axs[-1].set_xlabel("Time [s]", fontsize=12)
```
